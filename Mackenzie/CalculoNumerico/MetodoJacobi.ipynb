```python
#1.
#func para metodo jacobi
def MetodoJacobi(A, e:float):
  diff = 1000;
  tam = len(A)
  last_x = [0]*tam
  x_atual = [0]*tam


  while e < diff:

    for a in range(tam):
      div = A[a][a]
      if div == 0:
        continue
      elem = A[a][tam]/div
      for b in range(tam):
        if b == a:
          continue

        temp = (A[a][b]*last_x[b])/div
        elem -= temp

      x_atual[a] = elem

    nume = 0.0
    deno = 0.0
    for aux in range(tam):
      nume = max(abs(x_atual[aux] - last_x[aux]), nume)
      deno = max(abs(x_atual[aux]), deno)



    if deno == 0:
      diff = nume
    else:
      diff = nume/deno
    last_x = list(x_atual)

  return  x_atual
```

```python
#2. Exercicios Metodo de Gauss
#Se não rearranjarmos a matriz para a diagonal ficar certa o método nao
#consegue converger bem em um resultado
matriz_a = [
    [1, -1, 3, 2],
    [3, -3, 1, -1],
    [1, 1, 0, 3]
]

matriz_b = [
    [2, -1.5, 3, 1],
    [-1, 0, 2, 3],
    [4, -4.5, 5, 1]
]

matriz_c = [
    [2, 0, 0, 0, 3],
    [1, 1.5, 0, 0, 4.5],
    [0, -3, 0.5, 0, -6.6],
    [2, -2, 1, 1, 0.8]
]

matriz_d = [
    [1, 1, 0, 1, 2],
    [2, 1, -1, 1, 1],
    [4, -1, -2, 2, 0],
    [3, -1, -1, 2, -3]
]

e = 0.1
print()
print(f"Para e:{e}")
print("Matriz a:")
print(MetodoJacobi(matriz_a, e))
print("Matriz b:")
print(MetodoJacobi(matriz_b, e))
print("Matriz c:")
print(MetodoJacobi(matriz_c, e))
print("Matriz d:")
print(MetodoJacobi(matriz_d, e))

e = 0.01
print()
print(f"Para e:{e}")
print("Matriz a:")
print(MetodoJacobi(matriz_a, e))
print("Matriz b:")
print(MetodoJacobi(matriz_b, e))
print("Matriz c:")
print(MetodoJacobi(matriz_c, e))
print("Matriz d:")
print(MetodoJacobi(matriz_d, e))


e = 0.001
print()
print(f"Para e:{e}")
print("Matriz a:")
print(MetodoJacobi(matriz_a, e))
print("Matriz b:")
print(MetodoJacobi(matriz_b, e))
print("Matriz c:")
print(MetodoJacobi(matriz_c, e))
print("Matriz d:")
print(MetodoJacobi(matriz_d, e))


e = 0.0001
print()
print(f"Para e:{e}")
print("Matriz a:")
print(MetodoJacobi(matriz_a, e))
print("Matriz b:")
print(MetodoJacobi(matriz_b, e))
print("Matriz c:")
print(MetodoJacobi(matriz_c, e))
print("Matriz d:")
print(MetodoJacobi(matriz_d, e))
```
