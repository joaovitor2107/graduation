```python
import math
euler = math.e

def derivative(func, x, dx=1e-6):
    return (func(x + dx) - func(x - dx)) / (2 * dx)

def erro1(p1, p0):
  return abs(p1 - p0)

def erro2(p1, p0):
  return (abs(p1 - p0)/abs(p1))

def erro3(p1, f):
  return abs(f(p1))

def metodoNewton(f, p, e):
  p1 = p - (f(p)/derivative(f, p ,dx=1e-6))

  counte1 = 0
  counte2 = 0
  counte3 = 0
  flag1 = False
  flag2 = False
  flag3 = False

  while(erro1(p1, p) > e or erro2(p1, p) > e or erro3(p1, f) > e):
    p = p1
    p1 = p - (f(p)/derivative(f, p ,dx=1e-6))

    if(erro1(p1, p) <= e or flag1 == True):
      flag1 = True
    else:
      counte1 += 1
    if(erro2(p1, p) <= e or flag2 == True):
      flag2 = True
    else:
      counte2 += 1
    if(erro3(p1, f) <= e or flag3 == True):
      flag3 = True
    else:
      counte3 += 1

  print("Usando critério de parada 1: ", counte1, "iterações")
  print("Usando critério de parada 2: ", counte2, "iterações")
  print("Usando critério de parada 3: ", counte3, "iterações")
  print("Raiz encontrada: ", p1)


def fa(x):
  return (euler**x + 2**(-x) + 2*math.cos(x)- 6)

def fb(x):
  return (math.log(x - 1) + math.cos(x-1))

def fc(x):
  return (2*x*math.cos(2*x) - (x-2)**2)

def fd(x):
  return ((x-2)**2 - math.log(x))

def fe(x):
  return (euler**x - 3*x**2)

def ff(x):
  return (math.sin(x)-euler**(-x))

epsilon = 0.00001
print("Problema a")
metodoNewton(fa, (1+2)/2, epsilon)

print()
print("Problema b")
metodoNewton(fb, (1.3+2)/2, epsilon)

print()
print("Problema c")
metodoNewton(fc, (2+3)/2, epsilon)
metodoNewton(fc, (3+4)/2, epsilon)

print()
print("Problema d")
metodoNewton(fd, (1+2)/2, epsilon)
metodoNewton(fd, (euler+4)/2, epsilon)

print()
print("Problema e")
metodoNewton(fe, (0+1)/2, epsilon)
metodoNewton(fe, (3+5)/2, epsilon)

print()
print("Problema f")
metodoNewton(ff, (0+1)/2, epsilon)
metodoNewton(ff, (3+4)/2, epsilon)
```
