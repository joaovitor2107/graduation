```
def diferencasDivididas(nums, vals):
    size = len(nums)
    table = [[0.0 for _ in range(size)] for _ in range(size)]

    for i in range(size):
        table[i][0] = vals[i]

    for j in range(1,size):
        for i in range(size - j):
            table[i][j] = (table[i+1][j-1] - table[i][j-1])/(nums[i+j] - nums[i])

    coefi = [table[0][i] for i in range(size)]
    return coefi
```

```
n = int(input("Quantidade de numeros: "))
nums = []
vals = []

for _ in range(n):
   num = float(input("valor x: "))
   nums.append(num)
   val = float(input("valor f(x): "))
   vals.append(val)

print(diferencasDivididas(nums, vals))
```
